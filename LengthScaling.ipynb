```python
import numpy as np

# Scales window length to win_size
def LengthScaling(X, win_size=20):
    l = len(X)
    j = np.array(range(l))
    dj = np.floor(j / l * win_size)    # i is set to 0
    
    # Interpolation by j averaging --> Modify as needed
    sum_dj = np.zeros(win_size)
    num_dj = np.zeros(win_size)
    for _ in j:
        sum_dj[dj[_]] += X[_]
        num_dj[dj[_]] += 1
    return sum_dj / num_dj
```

```python
# Normalize to unit length: [0, 1]
# X must be ndarray
def Normalize(X):
    x_max = X.max(axis=0)
    x_min = X.min(axis=0)
    return (X - x_min) / (x_max - x_min)
```

```python
P1 = np.array([45, 47, 51, 49, 50, 51, 49, 53, 54, 55,
               57, 53, 54, 57, 56, 54, 57, 59, 61, 58,
               55, 57, 54, 53, 55, 57, 54, 51, 49, 51,
               48, 45, 47, 43, 44, 42, 47, 46, 49, 51])

import matplotlib.pyplot as plt
plt.plot(range(1, len(P1) + 1), P1, marker='o')
plt.xlabel('Time')
plt.ylabel('Price')
plt.show()
```

```python
P1_N = Normalize(P1)
print(P1_N)

plt.plot(range(1, len(P1_N) + 1), P1_N, marker='o')
plt.xlabel('Time')
plt.ylabel('Price_normalized')
plt.show()
```

```python
P1_N_S = LengthScaling(P1_N)
print(P1_N_S)

plt.plot(range(1, len(P1_N_S) + 1), P1_N_S, marker='o')
plt.xlabel('Time_scaled')
plt.ylabel('Price_normalized_scaled')
plt.show()
```
